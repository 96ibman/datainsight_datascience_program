```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
datasets.load_iris()
```

```python
def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df
```

# Query and filtering

```python
df = sklearn_to_df(datasets.load_iris())
df.head()
```

```python
df.query('`sepal length (cm)` > 7')
```

```python
myfilter = df['sepal length (cm)'] > 7
df[myfilter]
```

# Grouping and Aggregating
## Aggregation

```python
df = pd.DataFrame([[9, 4, 8, 9],
                   [8, 10, 7, 6],
                   [7, 6, 8, 5]],
                  columns=['Maths',  'English', 
                           'Science', 'History'])
  
print(df)
```

```python
df.sum()
```

```python
df.describe()
```

```python
df.agg(['sum', 'min', 'max'])
```

## Grouping

```python
df.groupby('Maths')
```

```python
a = df.groupby('Maths')
a.first()
```

```python
b = df.groupby(['Maths', 'Science'])
b.first()
```

# Apply

```python
df = pd.DataFrame([['Ahmed', 'Nasser', 90],
                   ['Ibrahim', 'John', 50],
                   ['Mohammed', 'Ali', 40]],
                  columns=['FirstName','SecondName',  'Average'])
  
print(df)
```

```python
def grading(num):
    if num <= 50:
        return 'Fail'
    else:
        return num
```

```python
df['Average'] = df['Average'].apply(grading)
```

```python
print(df)
```

# Dropping Duplicates

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df 
```

```python
df.drop_duplicates(subset=['brand'])
```

```python
df.drop_duplicates()
```
